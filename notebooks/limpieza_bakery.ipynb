```python
diccionario_semana = {"Monday" : "Lunes", "Tuesday": "Martes", "Wednesday":"Miércoles", "Thursday": "Jueves", "Friday": "Viernes", "Saturday": "Sábado", "Sunday":"Domingo"}

df["Día de la semana"] = df["Día de la semana"].replace(diccionario_semana)
```

```python
def estaciones(dato):

    primavera = np.union1d(pd.date_range(start= '2021-03-21', end='2021-06-21'), pd.date_range(start= '2022-03-21', end='2022-06-21'))
    verano = np.union1d(pd.date_range(start= '2021-06-22', end='2021-09-22'), pd.date_range(start= '2022-06-22', end='2022-09-22'))
    otoño = np.union1d(pd.date_range(start= '2021-09-23', end='2021-12-21'), pd.date_range(start= '2022-09-23', end='2022-12-21'))
    invierno = np.union1d(pd.date_range(start= '2021-12-22', end='2021-03-20'), pd.date_range(start= '2022-12-22', end='2022-03-20'))

    if dato in primavera:
        return 'Primavera'
    elif dato in verano:
        return 'Verano'
    elif dato in otoño:
        return 'Otoño'
    else:
        return 'Invierno'
```

```python
df['Estación'] = df['Fecha'].apply(estaciones)
df.sample(10)
```

```python
def franja(col):

    if int(col[:2]) < 12:
        return "Mañana"
    elif int(col[:2]) < 17:
        return "Mediodía"
    else:
        return "Tarde"
```

```python
df["Franja horaria"] = df["Hora"].apply(franja)
```
